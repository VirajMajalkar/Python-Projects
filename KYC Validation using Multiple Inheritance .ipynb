# KYC Validator Class With GUI

```python
from tkinter import *

gui = Tk() 
#text = Text(gui)
gui.configure(background = "light blue") 
gui.title("Aegis Institute of Data Science") 
gui.geometry("1200x750") 
text1 = StringVar()


lab1 = Label(gui,textvariable=text1,height = 2,width = 70,background = "light blue",anchor="w",font="Verdana", padx=10 )
lab1.grid(row=2,column=0)
text1.set("Select one of the KYC documents listed below : ")

lab12 = Label( gui,height = 1,width = 70, background = "light blue",anchor = "w",font = "Verdana", padx = 10 )
lab12.grid(row=1,column=0)

lab13 = Label( gui,height = 1,width = 70, background = "light blue",anchor = "w",font = "Verdana", padx = 10 )
lab13.grid(row=3,column=0)

lab14 = Label( gui,height = 1,width = 70, background = "light blue",anchor = "w",font = "Verdana", padx = 10 )
lab14.grid(row=6,column=0)

lab15 = Label( gui,height = 1,width = 70, background = "light blue",anchor = "w",font = "Verdana", padx = 10 )
lab15.grid(row=7,column=0)

lab16 = Label( gui,height = 1,width = 70, background = "light blue",anchor = "w",font = "Verdana", padx = 10 )
lab16.grid(row=10,column=0)

lab17 = Label( gui,height = 1,width = 70, background = "light blue",anchor = "w",font = "Verdana", padx = 10 )
lab17.grid(row=12,column=0)

lab18 = Label( gui,height = 1,width = 70, background = "light blue",anchor = "w",font = "Verdana", padx = 10 )
lab18.grid(row=14,column=0)



KYCDoc = StringVar()
var = IntVar()
choice = StringVar()
entryval1 = StringVar()
text10 = StringVar()
text7 = StringVar()
text11 = StringVar()


def sel():

    choice = var.get()   
    text2 = StringVar()
    
    lab2 = Label(gui,textvariable=text2,height=2,width=70,background = "light blue",anchor = "w",font="Verdana",padx=10)     
    lab2.grid(row=9,column=0)
     
    if choice == 1 :
        
        text2.set("Please Enter PAN : ")        
        
    elif choice == 2 :
        
        text2.set("Please Enter Passport Number : ")
            
    elif choice == 3 :
        
        text2.set("Please Enter Aadhar Number : ")    
    
    entrybox2 = Entry(gui,bd = 3,width = 15,font = "Vardana",textvariable = entryval1)
    entrybox2.grid(row = 9,column = 0)
    
    entryval1.set("")
    
    text10.set("")
    text11.set("")
    text7.set("")
    
    lab6 = Label(gui,textvariable=text11,height=1,width=70,background="light blue",anchor="w",font="Verdana",padx=10)
    lab6.grid(row=15,column=0)
    
    lab17 = Label(gui,textvariable=text11,height=1,width=70,background="light blue",anchor="w",font="Verdana",padx=10)
    lab17.grid(row=11,column=0)
              
    lab16 = Label (gui,textvariable=text11,height=1,width=70,background="light blue",anchor="w",font="Verdana",padx=10)
    lab16.grid(row=13,column=0)
    
    Button1 = Button (gui,text = "Validate",background = "Dark Blue",width = 15, font = ("Verdana",12,"bold"),foreground = "white",command = ValKYC.ValidateKYC)    
    Button1.grid(row = 9,column = 1)


R1 = Radiobutton(gui, text="PAN", background = "light blue",variable=var, value= 1,font = "Verdana",command =sel,anchor = "w",padx = 10)
R1.grid(row = 5,column = 0)

R2 = Radiobutton(gui, text="Passport", background = "light blue",variable=var, value= 2,font = "Verdana",command =sel, anchor = "w")
R2.grid(row = 5,column = 1)

R3 = Radiobutton(gui, text="Aadhar", background = "light blue",variable=var, value= 3,font = "Verdana",command =sel, anchor = "w",padx = 10)
R3.grid(row = 5,column = 2)

    
class ValPAN:
    
    def validatePAN(PAN):
        
        lab6 = Label( gui, textvariable=text7, height = 1,width = 70,background = "light green",anchor = "w",font = "Verdana",padx = 10 )    
        lab6.grid(row=15,column=0)

        if PAN.isalnum() == False:
                    
            text7.set("PAN Number contains non alpha numeric characters.")

        elif len(PAN) != 10:
            
            text7.set("PAN Number length is incorrect.")
        
        elif PAN[3] != 'P':
            
            text7.set("Entered PAN is of corporate customer.")
        
        else:
            
            text7.set("Entered PAN is valid PAN")

class ValPassport:
    
    def validatePassport(PassportNo):

        lab6 = Label( gui, textvariable = text7, height = 1,width = 70,background = "light green",anchor = "w",font = "Verdana",padx = 10 )    
        lab6.grid(row=15,column=0)

               
        if PassportNo.isalnum() == False:
            
            text7.set("Passport Number contains non alpha numeric characters")
            
        elif len(PassportNo) != 8:
            
            text7.set("Passport Number length is incorrect")

        elif PassportNo[0] != 'J':
            
            text7.set("Passport Number starts with J")
        
        else:
            
            text7.set("Entered Passport Number is valid")


class ValAadhar:

    def validateAadhar(AadharNo):
        
        lab6 = Label( gui, textvariable = text7, height = 1,width = 70,background = "light green",anchor = "w",font = "Verdana",padx = 10 )    
        lab6.grid(row=15,column=0)
                
        if AadharNo.isdigit() == False:
            
            text7.set("Aadhar Number contains non numeric characters")

        elif len(AadharNo) != 12:
            
            text7.set("Aadhar Number length is incorrect")

        elif AadharNo[0] == '0' or AadharNo[0] == '1' :
            
            text7.set("Aadhar Number does not start 0 or 1")
        
        else:
            
            text7.set("Entered Aadhar Number is valid")


class ValKYC(ValPAN,ValPassport,ValAadhar):
    
        
    def ValidateKYC():
        
        
        lab17 = Label( gui, textvariable = text11, height = 1,width = 70,background = "light yellow",anchor = "w",font = "Verdana",padx = 10 )    
        lab17.grid(row=11,column=0)
        
        text11.set("Calling ValidateKYC function inside Class ValKYC ")
        
        lab16 = Label( gui, textvariable = text10, height = 1,width = 70,background = "light yellow",anchor = "w",font = "Verdana",padx = 10 )    
        lab16.grid(row=13,column=0)

        choice = var.get()
        KYCDoc = choice
        entryvalue1 = entryval1.get()
                
        if KYCDoc == 1 :
            
            text10.set("calling ValidatePAN function inside class ValPAN")
            PAN = entryvalue1
            ValPAN.validatePAN(PAN)
        
        elif KYCDoc == 2 :
            
            text10.set("calling ValidatePassport function inside class ValPassport")
            PassportNo = entryvalue1
            ValPassport.validatePassport(PassportNo)
        
        elif KYCDoc == 3 :

            text10.set("calling ValidateAadhar function inside class ValAadhar")
            AadharNo = entryvalue1
            ValAadhar.validateAadhar(AadharNo)
        
        else:
        
            print("Incorrect KYC Doc entered")
    
#ValKYC.ValidateKYC('self')

gui.mainloop()
```
